----- ANS 1>>>  The fundamental difference is that DDL commands manage the database structure, DML commands manage the data within that structure, and DQL commands are used to retrieve that data.
DDL
DDL commands define or modify the database schema or structure. They deal with how the data is stored and organized, not the data itself. Changes made by DDL commands are typically auto-committed and permanent.
DML
DML commands are used to manipulate the data stored within the database tables. These operations are transactional, meaning they can be rolled back using TCL (Transaction Control Language) commands like COMMIT or ROLLBACK
DQL
DQL commands are used for querying and retrieving data from the database based on specific conditions. They do not modify the data or the database structure.

ANS 2>>> SQL constraints define rules to limit the data that can be stored in a table, ensuring the accuracy and integrity of the data within the database [1]. They can be applied to individual columns or the entire table.
NOT NULL - Ensures that a column cannot contain a NULL (empty) value [2]. This guarantees essential data is always present.
UNIQUE- Ensures that all values in a column are distinct [2]. While records with unique values can be present, no two entries can be identical.
PRIMARY KEY - A combination of NOT NULL and UNIQUE constraints [4]. It uniquely identifies each record in a table and ensures no duplicate rows or empty identifier values exist.

The SQL LIMIT clause restricts the maximum number of rows returned, while the OFFSET clause specifies the number of rows to skip from the beginning of the result set. They are commonly used together to implement pagination.
Difference Between LIMIT and OFFSET
LIMIT
* Function: Specifies the maximum number of rows that should be included in the result set.
* Usage: Determines the "page size"
* LIMIT 10 returns only the first 10 records.
OFFSET
* Function: Specifies how many rows to skip before starting to return the results.
* Usage: Determines the "starting point" of the data retrieval within the full result set.
* OFFSET 20 skips the first 20 records and starts returning from the 21st record onwards.

A Common Table Expression (CTE) is a temporary result set that you can reference within another SQL statement, such as a SELECT, INSERT, UPDATE, or DELETE statement [1]. It is defined using the WITH clause and exists only for the duration of the query's execution [1, 2].
   Main Benefits of Using CTEs.
* Readability and Maintenance: CTEs break down complex, lengthy queries into smaller, more manageable parts, making the SQL code much easier to read, understand, and maintain.
* Recursion: CTEs are capable of referencing themselves within their own definition, which allows for powerful recursive queries used to traverse hierarchical data such as organizational charts, file systems, or bill-of-materials structures .
  
  -- Define the CTE named 'CustomerTotals'
  
WITH CustomerTotals AS (
    -- This query calculates the total order amount per customer
    SELECT
        CustomerID,
        SUM(TotalAmount) AS TotalSpent
    FROM
        Orders
    GROUP BY
        CustomerID
)
-- Use the CTE in the main SELECT statement
SELECT
    c.CustomerName,
    ct.TotalSpent
FROM
    Customers c

WHERE
    ct.TotalSpent > 500;

SQL normalization is a systematic process of organizing a database to eliminate data redundancy and ensure data integrity.
PRIMARY GOALS OF NORMALIZATION.
* Reduce Data Redundancy: By storing data in a single, logical location, normalization avoids repetition of the same data across multiple tables.
* Improve Data Integrity: It helps maintain the accuracy and consistency of data, as changes need only be made in one place .
* Minimize Anomalies: Normalization helps prevent insertion, update, and deletion anomalies that can occur when data is not structured logically.

> The First Three Normal Forms (1NF, 2NF, and 3NF).
 first normal form.
 Each column contains atomic, or indivisible, values; a single cell should not hold multiple values .

 > Second Normal Form (2NF)
 It is already in 1NF.
 All non-key attributes (columns that are not part of the primary key) are fully dependent on the entire primary key [2]. If a table has a composite key, no non-key attribute can be dependent solely on only a part of that key.

 > Third Normal Form (3NF)
It is already in 2NF.
It has no transitive dependencies [2]. This means no non-key attribute should be dependent on another non-key attribute. All attributes must depend only on the primary key.

To generate a report showing the customer name, email, and total number of orders (including zero for those who haven't ordered), a LEFT JOIN is used between the customers table and the orders table. The COUNT() function will count the order_id, and by grouping by the customer, those without orders will show a count of 0.
 SQL QUEERY
 SELECT
    c.CustomerName,
    c.Email,
    COUNT(o.order_id) AS TotalNumberofOrders
FROM
    Customers c
LEFT JOIN
    Orders o ON c.CustomerID = o.CustomerID
GROUP BY
    c.CustomerID,
    c.CustomerName,
    c.Email
ORDER BY
    c.CustomerName;
    
    EXPLANATION
    
* SELECT c.CustomerName, c.Email: Selects the required columns from the customer table.
* COUNT(o.order_id): Counts only the matching orders for each customer. Because of the LEFT JOIN, customers with no orders will have a NULL order_id, which COUNT() treats as 0.
* FROM Customers c LEFT JOIN Orders o: Ensures all customers are included in the final report, even if they have no corresponding entries in the Orders table.
* GROUP BY c.CustomerID, ...: Groups the results to calculate the count specifically for each individual customer.
* ORDER BY c.CustomerName: Orders the resulting report alphabetically by the customer's name.

The SQL query to display the ProductName, Price, StockQuantity, and CategoryName for all products, ordered by CategoryName and then ProductName alphabetically, uses an INNER JOIN (or simply JOIN) to combine data from the Products and Categories.

SELECT
    Products.ProductName,
    Products.Price,
    Products.StockQuantity,
    Categories.CategoryName
FROM
    Products
INNER JOIN
    Categories ON Products.CategoryID = Categories.CategoryID
ORDER BY
    Categories.CategoryName,
    Products.ProductName;
    
    EXPLANATION
* SELECT Products.ProductName, ...: This specifies the columns to be retrieved from both the Products and Categories tables.
* FROM Products: This indicates the primary table for the query.
* INNER JOIN Categories ON Products.CategoryID = Categories.CategoryID: This joins the Products table with the Categories table based on the matching CategoryID in both tables. This ensures that only products with a valid, existing category are included in the results.
* ORDER BY Categories.CategoryName, Products.ProductName: This sorts the final result set first by CategoryName in ascending alphabetical order, and then by ProductName alphabetically for products within the same category.

You can retrieve the top two most expensive products for each category using a Common Table Expression (CTE) and the ROW_NUMBER() or DENSE_RANK() window function. The query assumes you have Products and Categories tables, linked by a CategoryID
WITH RankedProducts AS (
    SELECT
        C.CategoryName,
        P.ProductName,
        P.Price,
        ROW_NUMBER() OVER(PARTITION BY C.CategoryName ORDER BY P.Price DESC) AS RankNum
    FROM
        Products P
    JOIN
        Categories C ON P.CategoryID = C.CategoryID
)
SELECT
    CategoryName,
    ProductName,
    Price
FROM
    RankedProducts
WHERE
    RankNum <= 2
ORDER BY
    CategoryName,
    Price DESC;

EXPLANATION

* WITH RankedProducts AS (...): This defines a Common Table Expression (CTE) named RankedProducts to temporarily store a result set that includes a rank for each product.
* ROW_NUMBER() OVER(...): This window function assigns a sequential integer rank to each row within its partition.

PARTITION BY C.CategoryName: This divides the products into groups (partitions) by their category name.

ORDER BY P.Price DESC: This orders the products within each category by price in descending order (most expensive first).

* JOIN Categories C ON P.CategoryID = C.CategoryID: This joins the Products table with the Categories table to access the CategoryName.
* WHERE RankNum <= 2: The outer query filters the results from the CTE to only include products with a RankNum of 1 or 2, effectively selecting the top two most expensive products in each category.
* ORDER BY CategoryName, Price DESC: This sorts the final result set by category name and then by price in descending order for readability

    c.email,
    SUM(p.amount) AS total_spent

LIMIT 5;



2. Top 3 Movie Categories by Rental Count.
SELECT
    cat.name AS category_name,
    COUNT(r.rental_id) AS rental_count
FROM category cat
JOIN film_category fc ON cat.category_id = fc.category_id
JOIN film f ON fc.film_id = f.film_id
JOIN inventory i ON f.film_id = i.film_id
JOIN rental r ON i.inventory_id = r.inventory_id
GROUP BY cat.name
ORDER BY rental_count DESC
LIMIT 3;

3. Films Available per Store & Never Rented

SELECT
    s.store_id,
    COUNT(i.inventory_id) AS total_films,
    SUM(CASE WHEN r.rental_id IS NULL THEN 1 ELSE 0 END) AS never_rented
FROM store s
JOIN inventory i ON s.store_id = i.store_id
LEFT JOIN rental r ON i.inventory_id = r.inventory_id
GROUP BY s.store_id;

4. Monthly Revenue for 2023.
SELECT
    DATE_FORMAT(payment_date, '%Y-%m') AS month,
    SUM(amount) AS total_revenue
FROM payment
WHERE YEAR(payment_date) = 2023
GROUP BY month
ORDER BY month;

5. Customers with More Than 10 Rentals in Last 6 Months.
SELECT
    c.customer_id,
    CONCAT(c.first_name, ' ', c.last_name) AS customer_name,
    COUNT(r.rental_id) AS total_rentals
FROM customer c
JOIN rental r ON c.customer_id = r.customer_id
WHERE r.rental_date >= DATE_SUB(CURDATE(), INTERVAL 6 MONTH)
GROUP BY c.customer_id, customer_name
HAVING COUNT(r.rental_id) > 10;
